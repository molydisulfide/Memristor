```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
import scipy as sp
import csv
import matplotlib.patches as mpatches
import pylab as pb
import matplotlib.ticker as mtick
from matplotlib import cm
from numpy import linspace
from csv import reader
from matplotlib import pyplot
import matplotlib.colors as colors
import bokeh as bk
import pandas as pd
```

```python
width = []
HRS = []
LRS = []

with open('pulses.txt','r') as csvfile:
    plots = csv.reader(csvfile, delimiter='\t')
    for row in plots:
        width.append(float(row[0]))
        HRS.append(float(row[1]))
        LRS.append(float(row[2]))
       
        
      
```

```python
neg_beats = []
neg_time = []


with open('negative_beats.txt','r') as csvfile:
    plots = csv.reader(csvfile, delimiter='\t')
    for row in plots:
        neg_beats.append(float(row[0]))
        neg_time.append(float(row[1]))
        
pos_beats = []
pos_time = []


with open('positive_beats.txt','r') as csvfile:
    plots = csv.reader(csvfile, delimiter='\t')
    for row in plots:
        pos_beats.append(float(row[0]))
        pos_time.append(float(row[1]))
               
        
      
```

```python
x_Hz_01 = []
y_Hz_01 = []
x_Hz_1 = []
y_Hz_1 = []
x_Hz_10 = []
y_Hz_10 = []


with open('binary.txt','r') as csvfile:
    plots = csv.reader(csvfile, delimiter='\t')
    for row in plots:
        x_Hz_01.append(float(row[0]))
        y_Hz_01.append(float(row[1]))
        x_Hz_1.append(float(row[2]))
        y_Hz_1.append(float(row[3]))
        x_Hz_10.append(float(row[4]))
        y_Hz_10.append(float(row[5]))
```

```python
fig = plt.figure(figsize=(2.5,4.5))
fig.canvas.draw()
 
ax2 = fig.add_subplot(111)

plt.plot(width,HRS, '-s', ms = 8, lw = 2, color = 'mediumblue', mfc = 'white', mew = 0.5, 
         mec = 'mediumblue', label = 'HRS')
plt.plot(width,LRS, '-o', ms = 8, lw = 2, color = 'red', mfc = 'white', mew = 0.5, 
         mec = 'red', label = 'LRS')

  
#ax2.set_ylim([2e-9,8e-9])
ax2.set_xlim([1e-5,2])
ax2.set_xlabel('Pulse width (s)', fontsize = 12)
ax2.set_ylabel(' Read current after 5$^{th}$ pulse (A)', fontsize = 12)
plt.ticklabel_format(style='sci', axis='y')
plt.yticks(fontsize = 10)
plt.xticks(fontsize = 10)
ax2.set_title('V$_{read}$ = + 5 V')
ax2.set_xscale('log')
#ax2.set_yscale('log')

ax2.xaxis.set_ticks_position('bottom')
ax2.yaxis.set_ticks_position('left')
ax2.legend(loc = 'best', fontsize = 8, ncol =2)

plt.savefig("pulse_widths.png",bbox_inches='tight', dpi = 300)
plt.show()
```

```python
x = []
pre = []
neg = []
pos = []

with open('modulation1.txt','r') as csvfile:
    plots = csv.reader(csvfile, delimiter='\t')
    for row in plots:
        x.append(float(row[0]))
        pre.append(float(row[1]))
        neg.append(float(row[2]))
        pos.append(float(row[3]))
        

       
```

```python
fig = plt.figure(figsize=(6.5,6.5))
fig.canvas.draw()
 
ax2 = fig.add_subplot(111)

plt.plot(x[0:80],pre[0:80], '-', ms = 1, lw = 2, color = 'black', mew = 1, mec = 'mediumblue',
         label = 'Pre-modulation')
plt.plot(x[0:80],neg[0:80], '-', ms = 1, lw = 2, color = 'red', mew = 1, mec = 'mediumblue',
         label = 'Post -20 V modulation across V$_{12}$')
plt.plot(x[0:80],pos[0:80], '-', ms = 1, lw = 2, color = 'mediumblue', mew = 1, mec = 'mediumblue',
         label = 'Post +20 V modulation across V$_{12}$')


#ax2.set_ylim([0.1e-8,0.5e-8])
ax2.set_xlim([0,5])
ax2.set_xlabel('V$_{34}$ (V)', fontsize = 12)
ax2.set_ylabel(' I$_{34}$ (A) ', fontsize = 12)
plt.ticklabel_format(style='sci', axis='y')
plt.yticks(fontsize = 10)
plt.xticks(fontsize = 10)
#ax2.set_title('Alternating sign pulses every 0.01 s')
#ax2.set_xscale('log')
#ax2.set_yscale('log')

ax2.xaxis.set_ticks_position('bottom')
ax2.yaxis.set_ticks_position('left')
ax2.legend(loc = 'lower right', fontsize = 10, ncol =1)

plt.savefig("modulation1.png",bbox_inches='tight', dpi = 300)
plt.show()
```
