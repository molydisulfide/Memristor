```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
import scipy as sp
import csv
import matplotlib.patches as mpatches
import pylab as pb
import matplotlib.ticker as mtick
from matplotlib import cm
from numpy import linspace
from csv import reader
from matplotlib import pyplot
import matplotlib.colors as colors
import bokeh as bk
import pandas as pd
```

```python
bias = []
current = []
leakage = []

with open('leakygate.txt','r') as csvfile:
    plots = csv.reader(csvfile, delimiter='\t')
    for row in plots:
        bias.append(float(row[0]))
        current.append(float(row[1]))
        leakage.append(float(row[2]))

bias2 = []
current2 = []
leakage2 = []

with open('leakygate2.txt','r') as csvfile:
    plots = csv.reader(csvfile, delimiter='\t')
    for row in plots:
        bias2.append(float(row[0]))
        current2.append(float(row[1]))
        leakage2.append(float(row[2]))
        
gate = []
current3 = []
leakage3 = []

with open('gateleak.txt','r') as csvfile:
    plots = csv.reader(csvfile, delimiter='\t')
    for row in plots:
        gate.append(float(row[0]))
        current3.append(float(row[1]))
        leakage3.append(float(row[2]))
        
      
```

```python
bfielddc = []
vxydc = []


with open('magfielddc.txt','r') as csvfile:
    plots = csv.reader(csvfile, delimiter='\t')
    for row in plots:
        bfielddc.append(float(row[0]))
        vxydc.append(float(row[1]))
        
bfielddc_2 = []
vxydc_2 = []


with open('magfielddc_2.txt','r') as csvfile:
    plots = csv.reader(csvfile, delimiter='\t')
    for row in plots:
        bfielddc_2.append(float(row[0]))
        vxydc_2.append(float(row[1]))
        
lock_x = []
lock_y = []


with open('hall_lockin.txt','r') as csvfile:
    plots = csv.reader(csvfile, delimiter='\t')
    for row in plots:
        lock_x.append(float(row[0]))
        lock_y.append(float(row[1]))
```

```python
dc_fit = []
hall_fit = []


with open('hall_fit.txt','r') as csvfile:
    plots = csv.reader(csvfile, delimiter='\t')
    for row in plots:
        dc_fit.append(float(row[0]))
        hall_fit.append(float(row[1]))
        
dc_fit_2 = []
hall_fit_2 = []


with open('hall_fit_2.txt','r') as csvfile:
    plots = csv.reader(csvfile, delimiter='\t')
    for row in plots:
        dc_fit_2.append(float(row[0]))
        hall_fit_2.append(float(row[1]))
        
lock_x_fit = []
lock_y_fit = []


with open('hall_fit_lockin.txt','r') as csvfile:
    plots = csv.reader(csvfile, delimiter='\t')
    for row in plots:
        lock_x_fit.append(float(row[0]))
        lock_y_fit.append(float(row[1]))
        
```

```python
fig = plt.figure(figsize=(3.5,3.5))
fig.canvas.draw()
 
ax2 = fig.add_subplot(111)

plt.plot(bias,current, '-', ms = 1, lw = 1, color = 'forestgreen', label = 'I$_{ds}$ (A) ')
plt.plot(bias,leakage, '-', ms = 5, lw = 1, color = 'red', label = 'Gate Leakage')
  
ax2.set_ylim([5e-11,1e-6])
ax2.set_xlim([-20,20])
ax2.set_xlabel('V$_{ds}$ (V)', fontsize = 12)
ax2.set_ylabel(' I$_{ds}$ (A)', fontsize = 12)
plt.ticklabel_format(style='sci', axis='y')
plt.yticks(fontsize = 10)
plt.xticks(fontsize = 10)
ax2.set_title('Device #1')
#ax2.set_xscale('log')
ax2.set_yscale('log')

ax2.xaxis.set_ticks_position('bottom')
ax2.yaxis.set_ticks_position('left')
ax2.legend(loc = 'best', fontsize = 8, ncol =2)

plt.savefig("Hall_leaky.png",bbox_inches='tight', dpi = 300)
plt.show()
```
