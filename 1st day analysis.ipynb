```python
fig = plt.figure(figsize=(2.5,2.5))
fig.canvas.draw()
colormap = plt.get_cmap('coolwarm')

opened_file8 = open("All_gates_at_5V.txt")
csv_data8 = csv.reader(opened_file8, delimiter="\t")
parsed_data8 = []
for row in csv_data8:
    parsed_data8.append(row)
opened_file8.close()
data8 = np.array(parsed_data8)

ax4 = fig.add_subplot(111)

ax4.set_color_cycle([colormap(k) for k in np.linspace(0, 1, 15)])

for i in range(1,15):
        plt.plot(data8[0:100,0],data8[0:100,i], lw = 2)
        
ax4.set_ylim([1e-10,5e-5])
ax4.set_xlabel('Gate bias (V)', fontsize = 12)
ax4.set_ylabel('I$_{ds}$ (A)', fontsize = 12)
plt.ticklabel_format(style='sci', axis='y')
ax4.set_yscale('log')
plt.yticks(fontsize = 10)
plt.xticks(fontsize = 10)
#ax4.set_title('V$_{ds}$ = 5 V')
plt.setp(ax4.get_yticklabels(), visible=True)
plt.setp(ax4.get_xticklabels(), visible=True)
ax4.spines['right'].set_visible(False)
ax4.spines['top'].set_visible(False)
ax4.xaxis.set_ticks_position('bottom')
ax4.yaxis.set_ticks_position('left')

major_ticks_1 = np.arange(-60, 61, 60)   
ax4.set_xticks(major_ticks_1) 
plt.savefig("LRS_5V_gate_log.png",bbox_inches='tight', dpi = 300)
plt.show()
```
